```python
def impactPrice(price, amount):
    if amount <= 4 or price < 3:
        return min(price + np.random.randint(0,6), 99)
    elif amount <=40:
        return min(int(price*(1-0.000008*amount)) + np.random.randint(-3,6), 99)
    else:
        return min(int(98*(1- np.log(1831)*0.01)) + np.random.randint(-3,6), 99)
```

```python
price = [0 for i in range(20)]
price[0] = np.random.randint(1,99)
x = [0 for i in range(20)]
z = [0 for i in range(20)]
k = [0 for i in range(20)]
l = [0 for i in range(20)]
value = [0.0 for i in range(20)]
max_amount = 5000
max_price = 100
max_time = 20
amount = 4999
for time in range(20,0,-1):
    #print(time, price[max_time-time], amount)
    index = (time-1)*max_price*max_amount +(price[max_time-time])*max_amount+amount
    #print(index)
    value[max_time-time] = data[0][index]
    k[max_time-time] = data[1][index]
    l[max_time-time] = data[2][index]
    z[max_time-time] = int(amount*(2.73)**(l[max_time-time]*(time-21)))
    x[max_time-time] = int((amount-z[max_time-time])*(2.73)**(k[max_time-time]*(time-21)))
    amount = amount - x[max_time-time] - z[max_time-time]
    #print(x[max_time-time], z[max_time-time], k[max_time-time], l[max_time-time])
    if time > 1:
        price[max_time-time+1] = impactPrice(price[max_time-time], x[max_time-time])
```
